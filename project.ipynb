### Raw data 탐색 (null값 처리 및 불필요한 컬럼 제외)

```python
df = pd.read_csv('new_retail_data.csv')
df[:5]
```

```python
df.info()
```

```python
df_notnull = df.drop(columns=['Name', 'Email', 'Phone', 'Address'])
df_notnull[:5]
```

```python
print(df_notnull['Age'].dtype)
```

```python
cols = df_notnull.columns.to_list()
for col in cols:
    if df_notnull[col].dtype == 'float64':
        df_notnull[col].fillna(0.0, inplace=True)
    else:
        df_notnull[col].fillna('Unknown', inplace=True)
```

```python
df_notnull.info()
```

```python
df_notnull[df_notnull['City'] == 'Unknown']
```

```python
df_notnull = df_notnull.astype({'Transaction_ID': 'int', 'Customer_ID': 'int',
                                'Zipcode': 'int', 'Age': 'int', 'Year': 'int'})
df_notnull.dtypes
```

```python
# Transaction_ID: 거래 ID, Zipcode: 우편번호
df_notnull[:3]
```

```python
df_notnull.to_csv('./new_retail_data_notnull.csv', index=False)
```
